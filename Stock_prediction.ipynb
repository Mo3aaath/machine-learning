```python
import yfinance as yf
```

```python
sp500 = yf.Ticker('^GSPC')  # S&P 500 index
```

```python
sp500 = sp500.history(period='max') # making it into a pandas dataframe
```

```python
sp500
```

## Data cleaning

```python
# dropping Dividends and Stock Splits columns

sp500 = sp500.drop(["Dividends", "Stock Splits"], axis=1)
sp500
```

## Data visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure()
sns.lineplot(y="Close",x="Date", data=sp500)
plt.show()
```

## Target

```python
sp500["Tomorrow"] = sp500['Close'].shift(-1)
# sp500
```

```python
sp500["Target"] = (sp500['Tomorrow'] > sp500["Close"]).astype(int)
sp500
```

```python
sp500 = sp500.loc['1990-01-01':]
sp500
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
train = sp500[:-100]
test = sp500[-100:]

X = ["Open", "High", "Low", "Close"]
y = 'Target'

reg = LogisticRegression().fit(train[X], train[y])
```

## Precision

```python
from sklearn.metrics import precision_score
```

```python
test_pred = reg.predict(test[X])
```

```python
test_precision = precision_score(test[y], test_pred)
test_precision * 100
```
